```python
"""
Visual Studio Team Services
Python API samples

Click "Run" in the Jupyter menu to get started.

                    vSTs
                  vSTSVSTSv
                vSTSVSTSVST
 VSTS         vSTSVSTSVSTSV
 VSTSVS     vSTSVSTSV STSVS
 VSTSVSTSvsTSVSTSVS   TSVST
 VS  tSVSTSVSTSv      STSVS
 VS   tSVSTSVST       SVSTS
 VS tSVSTSVSTSVSts    VSTSV
 VSTSVST    SVSTSVSTs VSTSV
 VSTSv        STSVSTSVSTSVS
                VSTSVSTSVST
                  VSTSVSTs
                    VSTs    (TM)
"""

import sys
# tell python to look in .\src for loading modules
sys.path.insert(1, 'src')

from IPython.display import display
import ipywidgets as widgets

import runner
import runner_lib


def build_areas_and_resources():
    value = {
        'all': ['all',],
    }
    
    for area in runner_lib.discovered_samples.keys():
        value[area] = ['all',]
        value[area].extend(runner_lib.discovered_samples[area].keys())
    
    return value

areas_and_resources = build_areas_and_resources()


# build and display widgets
url_widget = widgets.Text(
    value='',
    placeholder='https://fabrikam.visualstudio.com',
    description='Account URL'
)
pat_widget = widgets.Text(
    value='',
    placeholder='VSTS PAT',
    description='PAT'
)
area_widget = widgets.ToggleButtons(
    options=areas_and_resources.keys(),
    description='Area'
)
resource_widget = widgets.ToggleButtons(
    options=areas_and_resources[area_widget.value],
    description='Resource'
)
run_widget = widgets.Button(
    description='Run samples',
    button_style='success',
    icon='play'
)

def on_area_change(change):
    resource_widget.options = areas_and_resources[change['new']]

area_widget.observe(on_area_change, names='value')

def on_run_click(b):
    if not url_widget.value or not pat_widget.value:
        print('You must specify a URL and PAT to run these samples.')
        return
    
    print('running samples...')
    print('-------------------------')
    print()
    runner.main(url=url_widget.value,
                auth_token=pat_widget.value,
                area=area_widget.value,
                resource=resource_widget.value
    )
    print()
    print('-------------------------')
    print('done!')

run_widget.on_click(on_run_click)

display(url_widget)
display(pat_widget)
display(area_widget)
display(resource_widget)
display(run_widget)
```
